```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.io


import time
import sys
import os.path
import joblib

import pywt

import cfl
```

```python
def wavMask(dims, scale):
    sx, sy = dims
    res = np.ones(dims)
    NM = np.round(np.log2(dims))
    for n in range(int(np.min(NM)-scale+2)//2):
        res[:int(np.round(2**(NM[0]-n))), :int(np.round(2**(NM[1]-n)))] = \
            res[:int(np.round(2**(NM[0]-n))), :int(np.round(2**(NM[1]-n)))]/2
    return res

def imshowWAV(Wim, scale=1):
    plt.imshow(np.abs(Wim)*wavMask(Wim.shape, scale), cmap = plt.get_cmap('gray'))

def coeffs2img(LL, coeffs):
    LH, HL, HH = coeffs
    return np.vstack((np.hstack((LL, LH)), np.hstack((HL, HH))))

def unstack_coeffs(Wim):
    L1, L2 = np.hsplit(Wim, 2) 
    LL, HL = np.vsplit(L1, 2)
    LH, HH = np.vsplit(L2, 2)
    return LL, [LH, HL, HH]

def img2coeffs(Wim, levels=3):
    LL, c = unstack_coeffs(Wim)
    coeffs = [c]
    for i in range(levels-1):
        LL, c = unstack_coeffs(LL)
        coeffs.insert(0,c)
    coeffs.insert(0, LL)
    return coeffs  

def dwt2(im):
    coeffs = pywt.wavedec2(im, wavelet='db4', mode='per', level=3)
    Wim, rest = coeffs[0], coeffs[1:]
    for levels in rest:
        Wim = coeffs2img(Wim, levels)
    return Wim

def idwt2(Wim):
    coeffs = img2coeffs(Wim, levels=3)
    return pywt.waverec2(coeffs, wavelet='db4', mode='per')

def fft2c(x):
    return 1 / np.sqrt(np.prod(x.shape[:2])) * np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(x, axes=(0, 1)), axes=(0, 1)), axes=(0, 1))

def ifft2c(y):
    return np.sqrt(np.prod(y.shape[:2])) * np.fft.ifftshift(np.fft.ifft2(np.fft.fftshift(y, axes=(0, 1)), axes=(0, 1)), axes=(0, 1))

def relnorm(x, y):
    return np.linalg.norm(x - y) / np.linalg.norm(x)
```

```python
data_dir = 'data'

teksp = cfl.readcfl(os.path.join(data_dir, 'teksp')).squeeze()
cfksp = cfl.readcfl(os.path.join(data_dir, 'cfksp')).squeeze()
sens = np.expand_dims(cfl.readcfl(os.path.join(data_dir, 'sens')).squeeze(), axis=3)
pat = np.expand_dims(cfl.readcfl(os.path.join(data_dir, 'mask')).squeeze(), axis=2)
bas = cfl.readcfl(os.path.join(data_dir, 'bas')).squeeze()
cfimg = np.expand_dims(cfl.readcfl(os.path.join(data_dir, 'cfimg_recon')).squeeze(), axis=2)

print ('teksp:', teksp.shape)
print ('cfksp:', cfksp.shape)
print ('sens:', sens.shape)
print ('pat:', pat.shape)
print ('bas:', bas.shape)
print ('cfimg:', cfimg.shape)

ny, nz, nc, T = teksp.shape
K_full = cfimg.shape[3]

Phi = bas[:, :K_full].real
```

## Scan and recon params

```python
model_dict = {
    'sens': sens,
    'pat': pat,
    'Phi': Phi,
}

def sens_forw(a, s):
    z = s * a
    return z

def sens_adj(a, s):
    z = np.expand_dims(np.sum(np.conj(s) * a, axis=2), 2)
    return z


def bas_forw(a, b):
    T, K = b.shape
    y_shape = list(a.shape)
    y_shape[-1] = T
    z = np.dot(b, a.reshape((-1, K)).T).T.reshape(y_shape)
    return z

def bas_adj(x, b):
    T, K = b.shape
    y_shape = list(x.shape)
    y_shape[-1] = K
    z = np.dot(np.conj(b.T), x.reshape((-1, T)).T).T.reshape(y_shape)
    return z


def fft_forw(x):
    return fft2c(x)

def fft_adj(y):
    return ifft2c(y)


def pat_forw(y, pat):
    return y * pat



def t2sh_forw(a, model_dict):
    sens = model_dict['sens']
    pat = model_dict['pat']
    Phi = model_dict['Phi']
    
    return pat_forw(bas_forw(fft_forw(sens_forw(a, sens)), Phi), pat)


def t2sh_adj(y, model_dict):
    sens = model_dict['sens']
    pat = model_dict['pat']
    Phi = model_dict['Phi']
    
    return sens_adj(fft_adj(bas_adj(pat_forw(y, pat), Phi)), sens)


def t2sh_normal(a, model_dict):
    return sens_adj(fft_adj(bas_adj(pat_forw(bas_forw(fft_forw(sens_forw(a, sens)), Phi), pat), Phi)), sens)

def stkern_mat(bas, mask):
    T, K = bas.shape
    ny, nz, _, nt = pat.shape
    assert(nt == T)
    z = pat.reshape((ny, nz, T, 1, 1)) * bas.reshape((1, 1, nt, 1, K))
    z2 = np.expand_dims(np.sum(z * bas.reshape((1, 1, T, K, 1)), axis=2), axis=2)
    return z2

def stkern_apply(a, stk):
    ny, nz, nc, K = a.shape
    z = np.sum(stk * a.reshape((ny, nz, nc, 1, K)), axis=4).reshape((ny, nz, nc, K))
    return z

def t2sh_normal_fast(a, model_dict):
    stk = model_dict['stkern']
    z1 = fft_forw(sens_forw(a, sens))
    z2 = stkern_apply(z1, stk)
    z3 = sens_adj(fft_adj(z2), sens)
    return z3
```

```python
cfimg = t2sh_adj(teksp, model_dict)


print('Computing space time kernel')
tic = time.time()
stk = stkern_mat(Phi, pat)
toc = time.time()
print('Done: {}'.format(toc - tic))
print()

model_dict['stkern'] = stk

print('Fast normal eq')
tic = time.time()
z1 = t2sh_normal_fast(cfimg, model_dict)
toc = time.time()
print('Done: {}'.format(toc - tic))
print()

print('Slow normal eq')
tic = time.time()
z2 = t2sh_normal(cfimg, model_dict)
toc = time.time()
print('Done: {}'.format(toc - tic))
print()

print('Error: {}'.format(relnorm(z1, z2)))
```

```python
def check_op(x_shape, y_shape, fn, an, x=None, y=None):
    if x is None:
        x = np.random.randn(*x_shape) + 1j * np.random.randn(*x_shape)
    if y is None:
        y = np.random.randn(*y_shape) + 1j * np.random.randn(*y_shape)
    Ax = fn(x)
    AHy = an(y)
    z1 = np.dot(np.conj(Ax.ravel()), y.ravel())
    z2 = np.dot(np.conj(x.ravel()), AHy.ravel())

    return np.linalg.norm(z1 - z2)


print(check_op((cfimg.shape), teksp.shape, lambda x: t2sh_forw(x, model_dict), lambda x: t2sh_adj(x, model_dict)))

cfcimg_shape = list(cfimg.shape)
cfcimg_shape[2] = sens.shape[2]
teimg_shape = list(cfimg.shape)
teimg_shape[3] = T

print(check_op(cfimg.shape, cfcimg_shape, lambda x: sens_forw(x, sens), lambda x: sens_adj(x, sens)))
print(check_op(cfimg.shape, teimg_shape, lambda x: bas_forw(x, Phi), lambda x: bas_adj(x, Phi)))
print(check_op(cfimg.shape, cfimg.shape, lambda x: fft_forw(x), lambda x: fft_adj(x)))
```

```python
def soft_thresh(x, lamb):
    res = np.abs(x) - lamb
    xhat = res * np.sign(x)
    xhat[res < 0] = 0.
    return xhat
    
def soft_thresh_cmplx(y, lamb):
    res = np.abs(y) - lamb
    res[res < 0] = 0
    xhat = res * np.exp(1j * np.angle(y))
    return xhat

def wave_soft_thresh_cmplx(x, lamb):
    z = idwt2(soft_thresh_cmplx(dwt2(x), lamb))
    return z

def wave_soft_thresh(x, lamb):
    z = idwt2(soft_thresh(dwt2(x), lamb))
    return z
```

```python
def reg_multi(fun_reg, cfimg, lamb):
    y = np.zeros(cfimg.shape, dtype=np.complex64)
    for i in range(cfimg.shape[-1]):
        y[:, :, 0, i] = fun_reg(cfimg[:, :, 0, i], lamb)
    return y

def fun_multi(fun, cfimg):
    y = np.zeros(cfimg.shape, dtype=np.complex64)
    for i in range(cfimg.shape[-1]):
        y[:, :, 0, i] = fun(cfimg[:, :, 0, i])
    return y
        
def zpad(cfimg):
    d0 = 264 - cfimg.shape[0]
    return np.pad(cfimg, ((d0//2, d0//2), (0, 0), (0, 0), (0, 0)), mode='wrap')

def crop(cfimg):
    d0 = cfimg.shape[0] - 260
    return cfimg[d0//2:d0//2+260,:,:,:]

def reg_multi_resize(fun_reg, cfimg, lamb):
    return crop(reg_multi(fun_reg, zpad(cfimg), lamb))

def fun_multi_resize(fun, cfimg):
    return crop(fun_multi(fun, zpad(cfimg)))
```

```python
import scipy.sparse.linalg
```

```python
def ist(fun_normal, fun_reg, lamb, x0, y_adj, params):
    nitr = params['nitr']
    step = params['step']
    callback = params['callback']
    x = x0.copy()
    
    
    for i in range(nitr):
        x  = fun_reg(x - step * (fun_normal(x) - y_adj), lamb * step)      
        callback(x)
        
    return x

def fista(fun_normal, fun_reg, lamb, x0, y_adj, params):
    nitr = params['nitr']
    step = params['step']
    callback = params['callback']
   
    t = 1.
    x = x0.copy()    
    
    for i in range(nitr):
        
        x_old = x.copy()
        x  = fun_reg(x - step * (fun_normal(x) - y_adj), lamb * step)

        t_old = t
        t = ((1 + np.sqrt(1 + 4 * t_old**2)) / 2.)
        
        x = x + (t_old - 1.) / t * (x - x_old)
        
        callback(x)
        
    return x


def conjgrad(fun_normal, fun_reg, lamb, x0, y_adj, params):
    nitr = params['nitr']
    callback = params['callback']
    
    ny, nz, _, nk = x0.shape
    
    assert fun_reg is None
   
    def mv(x):
        z = fun_normal(x.reshape((ny, nz, 1, nk)))
        return z.ravel()

    A_op = scipy.sparse.linalg.LinearOperator((ny*nz*nk, ny*nz*nk), matvec=mv, rmatvec=mv)
    x, _ = scipy.sparse.linalg.cg(A_op, y_adj.ravel(), x0=x0.ravel(), maxiter=nitr, callback=callback)
    
    return x.reshape((ny, nz, 1, nk))
```

```python
lamb = .01
nitr = 11
y_adj = t2sh_adj(teksp, model_dict)
z = y_adj.copy()
imgs = []
objval_ist = []
itr = 0

def callback(x):
    global itr
    global imgs
    global objval_ist
    f = np.linalg.norm(teksp.ravel() - t2sh_forw(x, model_dict).ravel())**2 * .5 + lamb * np.linalg.norm(fun_multi_resize(dwt2, x).ravel(), ord=1) 
    objval_ist.append(f)
    if itr % 5 == 0:
        print(itr)
        imgs.append(x)
        plt.figure()
        plt.imshow(np.abs(x.reshape((ny, nz, 1, K_full))[:,:,0,0]))

    itr = itr + 1
      
tic = time.time()
print('IST start')
z = ist(lambda x: t2sh_normal_fast(x, model_dict), 
          lambda x, y: reg_multi_resize(wave_soft_thresh_cmplx, x, y),
          lamb,
          z, y_adj, {'nitr': nitr, 'step': .95, 'callback': callback}
         )
toc = time.time()
print('Done: {} sec'.format(toc - tic))

plt.figure()
plt.plot(range(nitr), objval_ist)
plt.title('Objval')
```
